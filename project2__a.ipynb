
# **The dataset includes information about:**

Customers who left within the last month – the column is called Churn. Services that each customer has signed up for – phone, multiple lines, internet, online security, online backup, device protection, tech support, and streaming TV and movies. Customer account information – how long they had been a customer, contract, payment method, paperless billing, monthly charges, and total charges.Demographic info about customers – gender, age range, and if they have partners and dependents

```python
#importing the important projects 
import pandas as pd
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/ChurnData.csv")
display(df)
```

```python
#extracting the required parameters into another array
dff=df[["ed","employ","equip","callcard","wireless"]]
display(dff)
X = np.asanyarray(dff[["ed","employ","equip","callcard","wireless"]])
X[0:5]
y = np.asanyarray(df['churn'])
y [0:5]
```

```python
#preprocessing
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_t, X_test, y_t, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_t.shape,  y_t.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_t,y_t)
LR
```

```python
yhat = LR.predict(X_test)
yhat
```

```python
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

```python
a=confusion_matrix(y_test, yhat)
print(a)
```

```python
acc=a[0,0]+a[1,1]
acc1=a[0,0]+a[1,0]+a[1,1]+a[0,1]

print(acc/acc1)
```
